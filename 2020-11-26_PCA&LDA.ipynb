```python
%matplotlib inline
import matplotlib.pyplot as plt
from sklearn import datasets
from sklearn.decomposition import PCA
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
```

```python
data = datasets.load_iris()
data.feature_names
```

```python
x = data.data[:,:2]
y = data.target
target_names = data.target_names
```

```python
plt.figure(figsize=(10,10))
colors = ['red','blue']

for color, i, target_name in zip(colors,[0,1], target_names):
    plt.scatter(x[y==i,0],x[y==i,1],color=color, label = target_name)

plt.legend()
plt.xlabel('sepal length')
plt.ylabel('sepal width')
plt.show()
```

```python
x = data.data
y = data.target
target_names = data.target_names
```

```python
pca = PCA(n_components = 2)
x_p = pca.fit(x).transform(x)
print('가장 큰 주성분에 대한 분산: %s'%str(pca.explained_variance_ratio_))
```

```python
plt.figure(figsize = (10,10))
colors = ['red','blue']
for color, i, target_name in zip(colors, [0,1], target_names):
    plt.scatter(x_p[y ==i,0],x_p[y ==i,1], color = color, label = target_name)
    
plt.legend()
plt.xlabel('PC 1')
plt.ylabel('PC 2')
plt.show()
```
